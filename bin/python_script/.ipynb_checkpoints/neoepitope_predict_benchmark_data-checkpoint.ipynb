```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import datetime

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

from sklearn.feature_selection import RFE
```

```python
TME_in = False
```

```python
if TME_in == True: 
    cols_to_drop = ["Sample", "Patient","HLA_allele", "Peptide",'Allele_Frequency']
    TME = "included"
elif TME_in == False: 
    cols_to_drop = ["Sample", "Patient","HLA_allele", "Peptide",'Allele_Frequency','HLA_expression', 'CYT',
       'Monocytic lineage']
    TME = "excluded"
```

```python
# test with new data 


model_out_dir = '../../models/TME_'+TME + "_final"
rf_models = [filename for filename in os.listdir(model_out_dir) if filename.startswith("rf")]
print(model_out_dir)
#print(rf_models)

#pickled_model = pickle.load(open('model.pkl', 'rb'))
#pickled_model.predict(X_test)
```

```python
rf_total = list()
rf_models_0 = list(filter(lambda k: 'rf0' in k, rf_models))
rf_models_1 = list(filter(lambda k: 'rf1' in k, rf_models))
rf_models_2 = list(filter(lambda k: 'rf2' in k, rf_models))
rf_models_3 = list(filter(lambda k: 'rf3' in k, rf_models))
rf_models_4 = list(filter(lambda k: 'rf4' in k, rf_models))
rf_total = [rf_models_0,rf_models_1,rf_models_2,rf_models_3,rf_models_4]
```

```python
print(len(rf_total[1]))
```

```python
X_dat = pd.read_csv('../../data/03_partitioning_data/txt/03_1_1_EXTRA_pepides_for_analysis_for_model_V2.csv', sep='\t', low_memory=False)
```

```python
info = X_dat[['Sample', 'Patient','HLA_allele', 'Peptide']]
print(info.head)
print(X_dat.columns)
```

```python

partition = 0
extra_data  = pd.DataFrame()
pred_rf = 0
pred_df = pd.DataFrame()
pred_rf_total = pd.DataFrame()
pred_val_rf = 0
for rf in range(5):
    rf_models = rf_total[rf]
    print(rf_models)
    partition +=1
    av_pred_rf = 0
    num = 0
    for model in rf_models: 
        num += 1
        rf_model = pickle.load(open(model_out_dir +"/"+ model,"rb"))
       # print(rf_model)
        info = X_dat[["Sample", "Patient","HLA_allele", "Peptide"]]
        X_dat_drop = X_dat.drop(cols_to_drop, axis=1).reset_index(drop=True)
        predictions_rf_new = rf_model.predict_proba(X_dat_drop)
        av_pred_rf += predictions_rf_new
    print(num)
    pred_val_rf = av_pred_rf/num
    pred_rf = pd.DataFrame(pred_val_rf[:,1])
    pred_rf.reset_index(drop=True, inplace=True)
    X_dat.reset_index(drop=True, inplace=True)
    extra_data = pd.concat([X_dat,pred_rf],axis = 1)
    print(partition)
    extra_data["model"] = partition
       # print(bladder_results)
    pred_df = pred_df.append(extra_data)
```

```python
pred_df.to_csv(r'../../results/RandomForrest/extra_data/extra_data_prediction_update_20230221_'+TME+'.txt', index=None, sep='\t', mode='w')
```
