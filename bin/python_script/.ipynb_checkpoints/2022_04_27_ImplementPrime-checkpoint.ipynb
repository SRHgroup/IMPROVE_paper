```python
from past.utils import old_div
def rnaLogTrans(tpm,minVal = 1.0,maxVal=1500):
    #Apply normalization function as defined by Morten in some early publication 
    #- find that publication
    if tpm > maxVal:
        return 1.0
    elif tpm <= minVal:
        return 0.0#np.log10(minVal)
    else:
        return old_div(np.log10(tpm),np.log10(maxVal))#+np.log10(minVal)
    

dfMaster.sort_values('Gl',ascending=False)[['PeptMut','Gp','Gl','Core','CoreNonAnchor']].head()

dfMaster['ExprTrans'] = dfMaster['Expr'].apply(lambda tpm:rnaLogTrans(tpm,minVal=0.01))
dfMaster['ExprTrans'] = (dfMaster['ExprTrans']+abs(dfMaster['ExprTrans'].min()))/dfMaster['ExprTrans'].max()
dfMaster['ExprTrans'] = (dfMaster['ExprTrans'])/dfMaster['ExprTrans'].max()

dfMaster['Expr/EL_41'] = dfMaster['ExprTrans']/dfMaster['%Rank_EL_mut']
```

```python
import procTools
import pathTools
import shutil

def predReadWritePRIME(dfSel,predDir,utilsDir_prime,dataSet='BC',outFileTemplate = 'eval__PRIME__{}__{}.txt',resFileTemplate= 'prediction_{}_PRIME.tsv',tmpDir=False,**kwargs_prime):
    if tmpDir:
        ## This is a hack to get around the problem of PRIME not supporting filenames with spaces
        outDir = predDir        
        predDir=tmpDir
        pathTools.clearDirectory(tmpDir)
    else:
        outDir = predDir
    print("Run PRIME")
    print("Size of input: {}".format(len(dfSel)))
    procTools.groupAlleleRunProc(dfSel,predDir,predDir,peptCol='PeptMut',mhcCol='HLA',srcPath=utilsDir_prime,outFileTemplate = outFileTemplate,printCMD=False,**kwargs_prime)
    
    header_prime = ['Peptide','%Rank_bestAllele','Score_bestAllele','%RankBinding_bestAllele','BestAllele','%Rank_A0101','Score_A0101','%RankBinding_A0101']
    dfPrime = procTools.readNetOutDirApply(predDir,header=header_prime,cols=['Predictor','PeptType','HLA'],skiprows=12,skipfooter=0)
    dfPrime = dfPrime.rename(columns = {'%Rank_bestAllele':'%Rank_PRIME','Score_bestAllele':'Score_PRIME','Peptide':'PeptMut'})

    cols= ['PeptMut','%Rank_PRIME','Score_PRIME','HLA']
    dfPrime[cols].to_csv(os.path.join(outDir,'..',resFileTemplate.format(dataSet)),sep='\t',index=False)
    print("Size of PRIME output: {}".format(len(dfPrime)))
    if tmpDir:
        for file in os.listdir(tmpDir):
            shutil.copyfile(os.path.join(tmpDir,file), os.path.join(outDir,file))
    return dfPrime


tmpDir = '/Users/birkirreynisson/pac/tmp/'

utilsDir = '/Users/birkirreynisson/pac/utils/'
utilsDir_prime = os.path.join(utilsDir,'PRIME','PRIME')

resDir_prime_mix = os.path.join(resDir_prime,'res_mix')
#kwargs_prime = {'mix':'/Users/birkirreynisson/pac/utils/MixMHCpred/MixMHCpred'}
kwargs_prime = {}
dfPrime_mix = predReadWritePRIME(df,resDir_prime_mix,utilsDir_prime,dataSet = 'PRIME',tmpDir=tmpDir,**kwargs_prime)
```

# Generate information content dictionaries

```python
def getPseudoDict(path):
    pseudoDict = {}
    with open(path,'r') as fh:
        for line in fh:
            lineSplit = line.strip().split()
            pseudoDict[lineSplit[0]] = lineSplit[1]
    return pseudoDict

def getBGDict(path):
    bgDict = {}
    with open(path,'r') as fh:
        for line in fh:
            lineSplit = line.strip().split()
            bgDict[lineSplit[0]] = lineSplit[1]
    return bgDict

def readFreqMat(path):
    with open(path,'r') as fh:
        i = 0
        lineList = []
        for line in fh:
            line = line.strip()
            if line.startswith('#'):
                continue
            elif line.startswith('Last'):
                continue
            else:
                i+=1
            if i==1:
                aminos = line.split(' ')
            else:
                lineList.append(line.split(' ')[2:])
    return pd.DataFrame(lineList,columns=aminos)

def getFreqLOdfs(datDir,pseudoseq):
    dfFreq = readFreqMat(os.path.join(datDir,'fmat','{}_EL_100_freq.mat'.format(pseudoseq))).astype(float)
    dfLO = readFreqMat(os.path.join(datDir,'lomat','{}_EL_100.txt'.format(pseudoseq))).astype(float)
    return dfFreq,dfLO

def getInfoDF(dfFreq,dfLO):
    return (dfLO/2*dfFreq)

def getInfoSum(dfInfo):
    return dfInfo.sum(axis=1)

def infoSumWrapper(datDir,pseudoSeq):
    dfFreq,dfLO = getFreqLOdfs(datDir,pseudoSeq)
    dfInfo = getInfoDF(dfFreq,dfLO)
    infoSum = getInfoSum(dfInfo)
    return infoSum.values

def genInfoSumDict(datDir,hlas):
    pseudoDict = getPseudoDict(os.path.join(datDir,'MHC_pseudo.dat'))
    infoSumDict = {}
    for hla in hlas:
        hla = "HLA-{}".format(hla)
        infoSumDict[hla] = infoSumWrapper(datDir,pseudoDict[hla])
    return infoSumDict


def infoSumWrapper_prime(datDir,pseudoSeq):
    dfFreq = readFreqMat(os.path.join(datDir,'fmat','{}_EL_100_freq.mat'.format(pseudoSeq))).astype(float)
    dfInfo = getInfoDF_prime(dfFreq)
    infoSum = getInfoSum_prime(dfInfo)
    return infoSum.values

def getInfoDF_prime(dfFreq):
    return dfFreq*(np.log(dfFreq)/np.log(20))

def getInfoSum_prime(dfInfo):
    return 1+(dfInfo.sum(axis=1))

def genInfoSumDict_prime(datDir,hlas):
    pseudoDict = getPseudoDict(os.path.join(datDir,'MHC_pseudo.dat'))
    infoDict = {}
    for hla in hlas:
        pseudoSeq = pseudoDict["HLA-{}".format(hla)]
        infoDict[hla] = infoSumWrapper_prime(datDir,pseudoSeq)
    return infoDict

def infoDict2df(infoDict):
    return pd.DataFrame(infoDict).T.rename(columns= {i:i+1 for i in range(9)}).sort_index()

def dfInfo2dict(dfInfo):
    return dict(zip(dfInfo.index,dfInfo.values))

aminos='ARNDCQEGHILKMFPSTWYV'
hlas = list(set(list(dfMaster['HLA'].unique())+list(df['HLA'].unique())))
infoDict = genInfoSumDict_prime(datDir_logo,hlas)
dfInfo = infoDict2df(infoDict)
```

```python
def geo_mean_overflow(iterable):
    return np.exp(np.log(iterable).mean())

def posBoolDF(df,poses = list(range(4,9))):
    dfPos = df.copy(deep=True)

    for col in dfPos.columns:
        if col in poses:
            dfPos[col] = True
        else:
            dfPos[col] = False
    return dfPos

def threshBoolDF(df,thresh=0.3,gt=False):
    if gt:
        return df > thresh
    else:
        return df < thresh

def meanThreshBoolDF(df,gt=False,func = np.mean):
    if gt:
        return df.sub(df.apply(func,axis=1),axis=0) > 0
    else:
        return df.sub(df.apply(func,axis=1),axis=0) < 0

def and2boolDF(df1,df2):
    return (df1 & df2).astype(int)

#Naive: poses core[4:9]
dfInfoPos = posBoolDF(dfInfo,poses = list(range(4,9)))
dfWeight_pos = dfInfoPos.astype(int)
weightDict_pos = dfInfo2dict(dfWeight_pos)

#Positions in core[4:9] with InfoCont<0.3, i.e. like in PRIME
dfInfoPos = posBoolDF(dfInfo)
dfInfoThresh = threshBoolDF(dfInfo,thresh=0.20)
dfWeight_posThresh = and2boolDF(dfInfoPos,dfInfoThresh)
weightDict_posThresh = dfInfo2dict(dfWeight_posThresh)

#Positions with InfoCont lower than mean information content of motif
dfInfoPos = posBoolDF(dfInfo,poses = list(range(4,9)))
dfInfoThresh_mean = meanThreshBoolDF(dfInfo)
dfWeight_posMeanThresh = and2boolDF(dfInfoPos,dfInfoThresh_mean)
weightDict_mean = dfInfo2dict(dfWeight_posMeanThresh)

#Positions with InfoCont lower than mean information content of motif
dfInfoPos = posBoolDF(dfInfo,poses = list(range(4,9)))
dfInfoThresh_geoMean = meanThreshBoolDF(dfInfo,func=geo_mean_overflow)
dfWeight_posGeoMeanThresh = and2boolDF(dfInfoPos,dfInfoThresh_geoMean)
weightDict_geoMean = dfInfo2dict(dfWeight_posGeoMeanThresh)

#Weighted contribution, based on information content
dfWeight_info = dfInfo.sub(dfInfo.max(axis=1),axis=0).apply(abs)
weightDict_info = dfInfo2dict(dfWeight_info)
```

```python
dfInfoPos = posBoolDF(dfInfo,poses = list(range(4,9)))
dfInfoThresh_geoMean = meanThreshBoolDF(dfInfo,func=geo_mean_overflow)
dfWeight_posGeoMeanThresh = and2boolDF(dfInfoPos,dfInfoThresh_geoMean)
weightDict_geoMean = dfInfo2dict(dfWeight_posGeoMeanThresh)

dfWeight_info
dfInfoThresh_geoMean.astype(int)
dfWeight_posGeoMeanThreshInfo = dfWeight_posGeoMeanThresh*dfWeight_info
dfWeight_posGeoMeanThreshInfo = dfWeight_pos*dfWeight_info
weightDict_posInfo = dfInfo2dict(dfWeight_posGeoMeanThreshInfo)

dfWeight_posGeoMeanThreshInfoNorm = dfWeight_posGeoMeanThreshInfo.div(dfWeight_posGeoMeanThreshInfo.apply(max,axis=1),axis=0)
weightDict_posInfoNorm = dfInfo2dict(dfWeight_posGeoMeanThreshInfoNorm)
```

```python
def genWeightedEncoding(pept,encodingDict,posWeights):
    encDim = len(list(encodingDict.values())[0])
    encoding = np.zeros(encDim)
    for i,a in enumerate(pept):
        encoding+=encodingDict[a]*posWeights[i]
    return encoding

def addDeletion(pept,Gp,Gl):
    return pept[Gp:Gp+Gl]

def genMasterEncoding(df,encodingDict,weightDict,addDel=False,renameDict=False,addCols=['Target','Partition','Patient','HLA','PeptMut'],rankCol='%Rank_EL_mut'):
    enc = df.apply(lambda row: genWeightedEncoding(row['Core'],encodingDict,weightDict[row['HLA']]), axis=1)
    if type(addDel)==list:
        df['Del'] = df.apply(lambda row: addDeletion(row['PeptMut'],row['Gp'],row['Gl']) if row['Gp'] in addDel else '-',axis=1)
        enc+=df.apply(lambda row: genWeightedEncoding(row['Del'],encodingDict,len(row['Del'])*[1.0]), axis=1)        
    elif addDel:
        df['Del'] = df.apply(lambda row: addDeletion(row['PeptMut'],row['Gp'],row['Gl']) if row['Gl']>0 else '-',axis=1)
        enc+=df.apply(lambda row: genWeightedEncoding(row['Del'],encodingDict,len(row['Del'])*[1.0]), axis=1)
    dfEnc = pd.DataFrame(list(enc))
    dfEnc['pRank'] = df[rankCol].apply(lambda rank: -np.log10(rank+0.001))
    
    for col in addCols:
        dfEnc[col] = df[col]
    
    if renameDict:
        dfEnc = dfEnc.rename(columns = renameDict)
        #dfEnc = dfEnc[['Target','Partition','pRank']+list(renameDict.values())]
        dfEnc = dfEnc[addCols+['pRank']+list(renameDict.values())]
    features = dfEnc.columns[len(addCols):]
    featureDict = {a:True for a in features}
    return dfEnc,features,featureDict

def runCVtrainEval(dfEnc,features,featureDict,modelType='LogR',**modelKW):
    dfOut_logR, modelList = modelTrainEval.runCrossValidationTraining(dfEnc,features,featureDict,modelType=modelType,**modelKW)
    #NNanalysis.getAUC_EL_plotROC(dfOut_logR['Target'],dfOut_logR['classPred'],show=False)
    aucThresh = NNanalysis.getAUCThresh(dfOut_logR['Target'],dfOut_logR['classPred'])[-1]
    return aucThresh

def genEncTrainWrapper(df,encDict,weightDict,renameDict,addDel=False,modelType='LogR',**modelKW):
    dfEnc = genMasterEncoding(df,encDict,weightDict,addDel=addDel,renameDict = renameDict)
    features = dfEnc.columns[5:]
    featureDict = {a:True for a in features}
    aucThresh = runCVtrainEval(dfEnc,features,featureDict,modelType=modelType,**modelKW)
    return aucThresh

    
def genEncTrainWrapper_apply(df,hyperParams,hyperParams_str,modelType='LogR',**modelKW):
    hyperParamAUC = []
    for i,(weightDict,(encDict,renameDict),delHandle) in enumerate(hyperParams):    
        aucThresh = genEncTrainWrapper(df,encDict,weightDict,renameDict,addDel = delHandle,modelType=modelType,**modelKW)
        appendTup = (aucThresh,hyperParams_str[i])
        #print(appendTup)
        hyperParamAUC.append(appendTup)
    return hyperParamAUC
```
