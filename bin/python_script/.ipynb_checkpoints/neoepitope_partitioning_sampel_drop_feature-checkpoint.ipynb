```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import datetime

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn.neural_network import MLPClassifier
```

```python
# costumized stuff  
patient_data = True
```

```python
# run the data with or without the patient data for tumor microenviroment and with or withput selected cols
#print("running model with only_imp_cols:" + only_imp_cols + "and patient_data"+ patient_data)
cols_to_drop = ['Peptide', 'HLA_allele','Patient','Partition','cohort','est_freq']
#drop_col = ["Partition","response"]
model_data = pd.read_csv('../../data/03_partitioning_data/txt/03_2_filtered_data_model.txt', sep='\t')
#data = model_data [["mw","aro",'response','Peptide', 'HLA_allele','Patient','Partition','cohort','est_freq']]
n_sample = 10
```

```python
# nested cross validation
partitioning_list = model_data.Partition.unique()
pred_df_pr_feature =  pd.DataFrame()
AUC_values = pd.DataFrame()
#model_data = model_data[["mw","aro","partitoning","response"]]
count = 0
for column in model_data:
    count += 1
    if column in cols_to_drop:
        continue
    elif column=="response":
        drop_col = ['response','Peptide', 'HLA_allele','Patient','Partition','cohort','est_freq']
    else:
        drop_col = ['response','Peptide', 'HLA_allele','Patient','Partition','cohort','est_freq',column]
    pred_df = pd.DataFrame()
    data = model_data
    for i in partitioning_list:
        av_pred_rf = 0
        for n in range(n_sample):  
            test = data[data.Partition == i]
            train = data[data.Partition != i]
           # train_pos = train[train.response==1]
           # train_neg = train[train.response==0].sample(n=1000, random_state=n) #, random_state=45
           # train = shuffle(pd.concat([train_pos, train_neg], axis=0)).reset_index(drop=True)
            X_train = train.drop(drop_col, axis=1).reset_index(drop=True)
            y_train = train["response"]
            feature_list = list(X_train.columns)


            # Instantiate the RF and the MLP
            rf = RandomForestClassifier(n_estimators = 1000, random_state = 42, max_depth = 3)
            #rf = RandomForestClassifier(n_estimators = 1000, random_state = 42, max_depth = 3)
                    # Train the models on training data
            rf.fit(X_train, y_train)
            X_test = test.drop(drop_col, axis=1).reset_index(drop=True)
            y_test = test["response"]  
            test_cols = X_test.columns
            prediction_rf = rf.predict_proba(X_test)
            av_pred_rf += prediction_rf
        
        pred_val_rf = av_pred_rf/n_sample
        pred_val_rf  = pd.DataFrame(pred_val_rf [:,1])
        pred_val_rf.reset_index(drop=True, inplace=True)

        mapping = {pred_val_rf.columns[0]: column}
        pred_val_rf = pred_val_rf.rename(columns=mapping)
        X_test.reset_index(drop=True, inplace=True)
        y_test.reset_index(drop=True, inplace=True)
        ped_df = pd.concat([y_test,pred_val_rf],axis = 1)
        pred_df = pred_df.append(ped_df)

    pred_df_pr_feature = pd.concat([pred_df_pr_feature,pred_df[column]],axis = 1)
  #  AUC_values = [column,round(roc_auc_score(pred_df_pr_feature.response, pred_df_pr_feature.column),4)]
  #  print(round(roc_auc_score(pred_df_pr_feature.response, pred_df_pr_feature.column),4))
pred_df_pr_feature = pd.concat([pred_df_pr_feature,pred_df['response']],axis = 1)
```
