```python
# run the data with or without the patient data for tumor microenviroment and with or withput selected cols
#print("running model with only_imp_cols:" + only_imp_cols + "and patient_data"+ patient_data)
print(patient_data)
print(only_imp_cols)

if patient_data == True and only_imp_cols == False: 
    cols_to_drop = ['response','Mut_peptide', 'HLA_allele',"est_freq","grav","isopoint","cys_cys_bond"]#
    TME_inc = "TME_included"
if patient_data == False and only_imp_cols == False: 
    cols_to_drop = ['response','Mut_peptide', 'HLA_allele',"est_freq","grav","isopoint","cys_cys_bond","CYT","CD8A"]
    TME_inc = "TME_excluded"
    print(TME_inc)
if patient_data == True and only_imp_cols == True: 
    cols_to_drop = ['response','Mut_peptide', 'HLA_allele',"est_freq","grav","isopoint","cys_cys_bond","Expression_Level","Mut_MHCrank_EL"]
    TME_inc = "TME_included"
if patient_data == False and only_imp_cols == True: 
    cols_to_drop = ['response','Mut_peptide', 'HLA_allele',"est_freq","grav","isopoint","cys_cys_bond","CYT","CD8A","Expression_Level","Mut_MHCrank_EL","cellular_prevalence","variant_allele_frequency"]
    TME_inc = "TME_excluded"

    #
    

#my_auc=list()
n_sample = 10

pred_val_rf = []
pred_val_mlp = []

model_out_dir = "../../models/" +TME_inc + '_' +datetime.datetime.now().strftime('%Y-%m-%d_%H-%M-%S')
#os.makedirs("/home/xxx/" + datetime.datetime.now().strftime('%Y-%m-%d_%H-%M-%S'))
os.makedirs(model_out_dir)

Var_importance_filename = '../../results/RandomForrest/Feature_importance_' + TME_inc +'.txt'
outfile = open(Var_importance_filename ,'w')



test = pd.read_csv(Inputdir + '/f002', sep=' ', header=0,  names=names)
print(test)
```

```python

pred_df = pd.DataFrame()
for file in files:
    print("Partition:", file)
    filename = file[1:]
    #  train
    train = pd.read_csv(Inputdir + "/"+ file, sep=' ', header=0,  names=names)
    print(train.columns)
    train = train.dropna()
    train['response'].value_counts()
    
  #  if type(train["Self_Similarity"][0])==str:
  #      train = train[~train["Self_Similarity"].str.contains('acid')].reset_index(drop="True")
  #      train["Self_Similarity"] = train["Self_Similarity"].astype('float32')
        
        
    # test set 
    av_pred_rf = 0
    av_pred_mlp = 0
    
    test = pd.read_csv(Inputdir + '/c' + filename, sep=' ', header=0,  names=names)
    test = test.dropna()
    test['response'].value_counts()
    print(len(test))
   # test_pos = test[test.response==1]
    #  len(test_pos)
   # test_neg = test[test.response==0].sample(n=500, random_state=45)
   # test = shuffle(pd.concat([test_pos, test_neg], axis=0)).reset_index(drop=True)

        #Drop Self_similarity=='acid'
 #   if type(test["Self_Similarity"][0])==str:
 #       test = test[~test["Self_Similarity"].str.contains('acid')].reset_index(drop="True")
 #       test["Self_Similarity"] = test["Self_Similarity"].astype('float32')    
    
    # Because of downsampling is needed, make the downsampling 10 times (or 3 if you just debug :) ) 
    for i in range(n_sample): 
        train_pos = train[train.response==1]
        train_neg = train[train.response==0].sample(n=250, random_state=i) #, random_state=45
        train = shuffle(pd.concat([train_pos, train_neg], axis=0)).reset_index(drop=True)
        X_train = train.drop(cols_to_drop, axis=1).reset_index(drop=True)
        y_train = train["response"]
        
        feature_list = list(X_train.columns)
        
        # Instantiate the RF and the MLP
        rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
        mlp = MLPClassifier(max_iter=500, hidden_layer_sizes=200)
    
        # Train the models on training data
        rf.fit(X_train, y_train)
        mlp.fit(X_train, y_train)
        filename = model_out_dir +"/rf_"+file+"_"+"sample"+str(i)+".sav"
        pickle.dump(rf,open(filename,"wb"))
     #   pickle.dump(mlp,open("../../models/Version1/mlp_"+file+"sample"+str(i)),"wb")
        
        #Get fetures importance per partition
        importances = list(rf.feature_importances_)
        feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]
        # Sort
        feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
        [print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances]
        # gather all feature importance 
        for pair in feature_importances:
            outfile.write(str(pair[0]) + "\t" + str(pair[1]) + "\n")
            
     

        # Test set 
        test.columns = train.columns 
        X_test = test.drop(cols_to_drop, axis=1).reset_index(drop=True)
        y_test = test["response"]  
        test_cols = X_test.columns
     
    # Plot the importances 
    #     plt.style.use('fivethirtyeight')
    #     x_values = list(range(len(importances)))
    #     plt.bar(x_values, importances, orientation = 'vertical')
    #     plt.xticks(x_values, feature_list, rotation='vertical')
    #     plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances')
    #     plt.show()

    # Predict values 
    
        predictions_rf = rf.predict_proba(X_test)
        predictions_mlp = mlp.predict_proba(X_test)
        
        av_pred_rf += predictions_rf
        av_pred_mlp += predictions_mlp
    
    pred_val_rf = av_pred_rf/n_sample
    pred_val_mlp = av_pred_mlp/n_sample
    
    pred_df = pred_df.append(pd.concat([pd.DataFrame(X_test, columns=X_test.columns), 
    pd.Series(y_test, name='response'), 
    pd.Series(pred_val_rf[:,1], name='prediction_rf'),                          
    pd.Series(pred_val_mlp[:,1], name='prediction_mlp')], axis=1)).reset_index(drop=True)

pred_df=pred_df.dropna()
print(pred_df)
# Print final AUCs 
print("RF AUC:",round(roc_auc_score(pred_df.response, pred_df.prediction_rf),4))
print("MLP AUC:",round(roc_auc_score(pred_df.response, pred_df.prediction_mlp),4))
    
            
        
outfile.close()        
        

        
        
```

```python
plt.hist(pred_df.prediction_rf, alpha=0.5, label='prediction')
plt.hist(pred_df.response, alpha=0.5, label='response')
plt.legend()
plt.title("Histogram of the responses and the predicted probabilities\n Ranfom Forest")
plt.show()

plt.hist(pred_df.prediction_mlp, alpha=0.5, label='prediction')
plt.hist(pred_df.response, alpha=0.5, label='response')
plt.legend()
plt.title("Histogram of the responses and the predicted probabilities\n MLP")
plt.show()
```

```python
fpr_rf, tpr_rf, threshold_rf = roc_curve(pred_df.response, pred_df.prediction_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

fpr_mlp, tpr_mlp, threshold_mlp = roc_curve(pred_df.response, pred_df.prediction_mlp)
roc_auc_mlp = auc(fpr_mlp, tpr_mlp)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr_rf, tpr_rf, 'b', label = 'AUC RF= %0.2f' % roc_auc_rf)
plt.plot(fpr_mlp, tpr_mlp, 'g', label = 'AUC MLP= %0.2f' % roc_auc_mlp)

plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
