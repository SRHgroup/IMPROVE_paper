```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.layers import Input, Dense
from keras.models import Sequential, Model
from sklearn.metrics import roc_auc_score, mean_squared_error
from sklearn.utils import shuffle
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
data_dir = "../../data/partitioning_data/From_computerome/"

names = ["Mut_peptide", "HLA_allele", "cellular_prevalence", "Expression_Level", 
         "priority_Score", "Allele_Frequency", "Mut_MHCrank_EL", "Mut_MHCrank_BA", 
         "variant_allele_frequency", "response", "Self_Similarity", "exp_rank_binary", 
         "clonal_mut", "hydro", "pI", "Cancer_Driver_Gene", "Mutation_Consequence", "est_freq"]
```

```python
#Declare the model
def simple_ann(in_shape):
    data_in = Input(shape=in_shape)
    dense1 = Dense(15, activation='relu')(data_in)
    out1 = Dense(1, activation='sigmoid')(dense1)
    out2 = Dense(1, activation='linear')(dense1)
    
    return Model(inputs=data_in, outputs=[out1,out2])


#Declare the model One Output node
def simple_ann_one(in_shape):
    data_in = Input(shape=in_shape)
    dense1 = Dense(15, activation='relu')(data_in)
    out1 = Dense(1, activation='sigmoid')(dense1)
    
    return Model(inputs=data_in, outputs=[out1])
    
```

```python
pred_df = pd.DataFrame()
pred_df_one = pd.DataFrame()
for n in range(5):
    print("Fold", n)
    train_data = pd.read_csv(data_dir + 'f00' + str(n), names=names, sep=' ', header=0)
    test_data = pd.read_csv(data_dir + 'c00' + str(n), names=names, sep=' ', header=0)
    
    #Drop NaN
    train_data = train_data.dropna().reset_index(drop=True)
    test_data = test_data.dropna().reset_index(drop=True)
    
    # downsample data 
    train_pos = train_data[train_data.response==1]
    train_neg = train_data[train_data.response==0].sample(n=400, random_state=45)
    train_data = shuffle(pd.concat([train_pos, train_neg], axis=0)).reset_index(drop=True)
    
    
    test_pos = test_data[test_data.response==1]
    test_neg = test_data[test_data.response==0].sample(n=100, random_state=45)
    test_data = shuffle(pd.concat([test_pos, test_neg], axis=0)).reset_index(drop=True)


    #Drop Self_similarity=='acid'
    if type(train_data["Self_Similarity"][0])==str:
        train_data = train_data[~train_data["Self_Similarity"].str.contains('acid')].reset_index(drop="True")
        train_data["Self_Similarity"] = train_data["Self_Similarity"].astype('float32')
    if type(test_data["Self_Similarity"][0])==str:
        test_data = test_data[~test_data["Self_Similarity"].str.contains('acid')].reset_index(drop="True")
        test_data["Self_Similarity"] = test_data["Self_Similarity"].astype('float32')
    
    
    # Two output node 
    #Drop the target variable and define y
    y_train = train_data[["response", "est_freq"]]
    y_test = test_data[["response", "est_freq"]]
    
    X_train = train_data.drop(columns=['response','Mut_peptide','HLA_allele','est_freq'])
    X_test = test_data.drop(columns=['response','Mut_peptide','HLA_allele','est_freq'])
    
    #train
    model = simple_ann(in_shape=14)
    
    model.compile(loss=['binary_crossentropy','mean_squared_error'], optimizer='adam')
    model.fit(X_train, [y_train.response, y_train.est_freq], validation_split=0.2, epochs=100, batch_size=64, verbose=0)
    
    pred = model.predict(X_test)
    
    
    pred_df = pred_df.append(pd.concat([pd.DataFrame(test_data),
                                        pd.Series(np.ones(test_data.shape[0])*n, name="partition"),
                                        pd.Series(np.squeeze(pred[0]), name='response_pred'),
                                        pd.Series(np.squeeze(pred[1]), name='est_freq_pred')], axis=1)).reset_index(drop=True)
    
    
    #One output node 
    y_train_one = train_data[["response"]]
    y_test_one = test_data[["response"]]
    
    
    model_one = simple_ann_one(in_shape=14)
    
    model_one.compile(loss=['binary_crossentropy','mean_squared_error'], optimizer='adam')
    model_one.fit(X_train, [y_train_one.response], validation_split=0.2, epochs=100, batch_size=64, verbose=0)
    
    pred_one = model_one.predict(X_test)
    
    
    pred_df_one = pred_df_one.append(pd.concat([pd.DataFrame(test_data),
                                        pd.Series(np.ones(test_data.shape[0])*n, name="partition"),
                                        pd.Series(np.squeeze(pred_one), name='response_pred')], axis=1)).reset_index(drop=True)
    
    
  #  pred_df_one = pred_df_one.append(pd.concat([pd.DataFrame(X_test, columns=X_test.columns), 
   #                                     pd.Series(y_test_one, name='response'), 
    #                                    pd.Series(pred_one, name='response_pred')], axis=1)).reset_index(drop=True)
                                     

    
    
```

```python
# performance two output node 
print("Test AUC on response: {:.4f}".format(roc_auc_score(pred_df.response, pred_df.response_pred)))
print("MSE on est_freq: {:.4f}".format(1/pred_df.shape[0]*(np.sum((pred_df.est_freq-pred_df.est_freq_pred)**2))))
```

```python
# performance one output node 
print("Test AUC on response: {:.4f}".format(roc_auc_score(pred_df_one.response, pred_df_one.response_pred)))
```

```python
# roc curves 
fpr_rf, tpr_rf, threshold_rf = roc_curve(pred_df.response, pred_df.response_pred)
roc_auc_2_out = auc(fpr_rf, tpr_rf)

fpr_mlp, tpr_mlp, threshold_mlp = roc_curve(pred_df_one.response, pred_df_one.response_pred)
roc_auc_1_out = auc(fpr_mlp, tpr_mlp)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr_rf, tpr_rf, 'b', label = 'AUC 2 Out= %0.2f' % roc_auc_2_out)
plt.plot(fpr_mlp, tpr_mlp, 'g', label = 'AUC 1 Out= %0.2f' % roc_auc_1_out)

plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
