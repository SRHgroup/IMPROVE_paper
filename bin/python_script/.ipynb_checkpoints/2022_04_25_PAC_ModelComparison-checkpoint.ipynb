### RNA TPM logtransform

```python
def rnaLogTrans(tpm,minVal = 1.0,maxVal=1500):
    #Apply normalization function as defined by Morten in some early publication 
    #- find that publication
    if tpm > maxVal:
        return 1.0
    elif tpm <= minVal:
        return 0.0#np.log10(minVal)
    else:
        return old_div(np.log10(tpm),np.log10(maxVal))#+np.log10(minVal)
    

dfMaster.sort_values('Gl',ascending=False)[['PeptMut','Gp','Gl','Core','CoreNonAnchor']].head()

dfMaster['ExprTrans'] = dfMaster['Expr'].apply(lambda tpm:rnaLogTrans(tpm,minVal=0.01))
dfMaster['ExprTrans'] = (dfMaster['ExprTrans']+abs(dfMaster['ExprTrans'].min()))/dfMaster['ExprTrans'].max()
dfMaster['ExprTrans'] = (dfMaster['ExprTrans'])/dfMaster['ExprTrans'].max()

dfMaster['Expr/EL_41'] = dfMaster['ExprTrans']/dfMaster['%Rank_EL_mut']
```

### Gen CYT scores from RNA

```python
def addDScol(sample):
    if 'MM909' in sample:
        return 'Neye'
    if 'PATIEN' in sample:
        return 'RH'
    else:
        return 'BC'

def procSampleCol(sample):
    sample = sample.replace('MM909_','')
    sample = sample.replace('PATIEN','')
    sample = sample.replace('A','')
    sample = sample.replace('B','')
    sample = sample.split('_')[0]
    sample = int(sample)
    return sample

def getHouseKeepingRNA(df,houseKeepingGenes=[]):
    df = df.rename(columns={'hugo_symbol':'Gene','mean_exp':'Exp'})
    if houseKeepingGenes:
        dfFilt = df[df['Gene'].apply(lambda gene: gene in houseKeepingGenes)]
    else:
        dfFilt = df
    dfFilt.loc[:,'Exp'] = dfFilt['Exp'].apply(lambda exp: round(float(exp),2))
    dfFilt.loc[:,'Sample'] = dfFilt['Sample'].apply(str)
    dfFilt.loc[:,'Dataset'] = dfFilt['Sample'].apply(addDScol)
    dfFilt.loc[:,'Sample'] = dfFilt['Sample'].apply(procSampleCol)
    dfFilt.loc[:,'Patient'] = dfFilt.apply(lambda row: "{}-{}".format(row['Dataset'],row['Sample']),axis=1)
    dfFilt = dfFilt.groupby(['Dataset','Sample','Gene','Patient'])['Exp'].agg(np.median).reset_index()
    return dfFilt

def geo_mean_overflow(iterable):
    return np.exp(np.log(iterable).mean())

dfRNAfilt_cyt = getHouseKeepingRNA(dfRNA,houseKeepingGenes=['GZMA','PRF1'])
dfRNAfilt_cyt_pivot = dfRNAfilt_cyt[['Gene','Patient','Exp']].pivot(index='Gene',columns='Patient',values='Exp')
dfRNAfilt_cyt_pivot = dfRNAfilt_cyt_pivot.T.reset_index()

dfRNAfilt_cyt_pivot['CYT'] = dfRNAfilt_cyt_pivot.apply(lambda row: geo_mean_overflow([row['GZMA'],row['PRF1']]),axis=1)

dfRespCount = dfMaster.groupby(['Dataset','Sample','Patient'])['Target'].apply(sum).reset_index().rename(columns={'Target':'Detected'})
dfRespCount = dfRespCount.merge(dfRNAfilt_cyt_pivot,on='Patient')
print(len(dfRespCount))
#dfRespCountDTU

print(len(dfMaster))
dfMaster_cyt = dfMaster.merge(dfRespCount[['Patient','CYT']].drop_duplicates(),on='Patient')
print(len(dfMaster_cyt))
```

```python
def getMultiplcityDistribution(df,coveredAlleles,predCol='meanPred',rank=False):
    if coveredAlleles:
        dfHLA = dfTools.dfColContainsAnyFilter(df,coveredAlleles,'HLA')
    else:
        dfHLA = df
    dfHLACount = dfHLA.groupby('Patient')['HLA'].nunique().reset_index().rename(columns={'HLA':'HLA-Hits'})
    dfHLA['Dataset']  = dfHLA['Patient'].apply(lambda pat: pat.split('-')[0])
    dfHLA_T = dfHLA[dfHLA['Dataset']=='Tesla']
    dfHLATop50_T = rankEpiTools.getGroupTopX(dfHLA_T,predCol=predCol,rank=rank,X=20)
    dfHLA_NT = dfHLA[dfHLA['Dataset']!='Tesla']    
    dfHLATop50_NT = rankEpiTools.getGroupTopX(dfHLA_NT,predCol=predCol,rank=rank)
    dfHLATop50 = pd.concat([dfHLATop50_NT,dfHLATop50_T])
    #dfHLATop50 = rankEpiTools.getGroupTopX(dfHLA,predCol='meanPred',rank=False)
    dfHLATop50HitCount = dfHLATop50.groupby('Patient')['Target'].sum().reset_index().rename(columns={'Target':'Top50-Hits'})
    dfHLATop50HitCount = dfHLATop50HitCount.merge(dfHLACount)
    dfHLATop50HitCount['Top50-Hits'] = dfHLATop50HitCount['Top50-Hits'].apply(int)
    return dfHLATop50HitCount

def getMultiplicityRatio(df,multiplicityThreshold=2,hlaHitThresh=2):
    df = df[df['Top50-Hits']>0]
    df = df[df['HLA-Hits']>=hlaHitThresh]
    return np.mean(df['Top50-Hits']>=multiplicityThreshold)

def multiplicityHistograms(df,x='Top50-Hits',multiplicityThreshold=2,threshes=range(1,3),saveFig=False,color='dodgerblue'):
    df = df[df[x]>0]
    for hlaThresh in threshes:
        dfOut = df[df['HLA-Hits']>=hlaThresh]

        maxx = int(dfOut[x].max())
        maxy = dfOut.groupby(x).apply(len).max()

        mean = round(dfOut[x].mean(),2)
        median = int(dfOut[x].median())
        multiPlicityRatio = getMultiplicityRatio(dfOut,hlaHitThresh=hlaThresh,multiplicityThreshold=multiplicityThreshold)
        multiplicity = int(multiPlicityRatio*100)

        plt.figure(figsize=(12,8))
        
        plt.text(0.75*maxx,0.90*maxy,"Mean: {}".format(mean),color=sns.color_palette()[0])  
        plt.text(0.75*maxx,0.85*maxy,"Median: {}".format(median),color=sns.color_palette()[1])
        plt.text(0.75*maxx,0.80*maxy,"Multiplicity: {}%".format(multiplicity),color=sns.color_palette()[3])        
        #ratio = 0.6
        
        plt.axvline(mean,linewidth=4.0,color=sns.color_palette()[0])
        plt.axvline(median,linewidth=4.0,color=sns.color_palette()[1])
        #splot = sns.distplot(dfOut['Top50-Hits'],kde=False,bins=maxx-1,color='dodgerblue')
        splot = sns.distplot(dfOut['Top50-Hits']-0.35,kde=False,bins=maxx-1,hist_kws={"histtype": "bar",'width':0.7, "linewidth": 3,"alpha": 0.6, "color": color})        
        #splot = sns.countplot(data=dfOut,x=x,color='dodgerblue')
        
        plt.xlabel('Epitopes in Predicted Top 50')
        plt.ylabel('Number of Patients')
        plt.xlim((0,maxx))
        plt.xticks(range(0,maxx+1))
        plt.yticks(range(0,maxy+1,2))
        if saveFig:
            plt.savefig(saveFig.format(hlaThresh),dpi=600)
        plt.show()

coveredAlleles = ['A0101','A0201','A0301','A1101','A2402','B0702','B0801']
coveredAlleles = False
saveFig = os.path.join(figDir,"multiplicityDistribution_7HLAs_HLAhits{}.png")
dfCVEval_seed_RF_BC= dfCVEval_seed_RF[dfCVEval_seed_RF['Dataset']=='BC']
#dfCVEval_seed_RF_BC_noC = dfCVEval_seed_RF_BC[dfCVEval_seed_RF_BC['Loci']!='C']
dfCVEval_seed_RF_BC_noC = dfCVEval_seed_RF_BC
dfCVEval_seed_RF_hla_top50_hitCount_hlaCount = getMultiplcityDistribution(dfCVEval_seed_RF_BC_noC,coveredAlleles,predCol='meanPred')
#dfCVEval_seed_RF_hla_top50_hitCount_hlaCount = getMultiplcityDistribution(dfCVEval_seed_RF,coveredAlleles,predCol='%Rank_EL_mut',rank=True)
#dfCVEval_seed_RF_hla_top50_hitCount_hlaCount = getMultiplcityDistribution(dfCVEval_seed_RF,coveredAlleles,predCol='%Rank_PRIME',rank=True)
multiplicityHistograms(dfCVEval_seed_RF_hla_top50_hitCount_hlaCount,threshes = range(0,1),saveFig=False,color=sns.color_palette()[2])
```
